```python
import os
import time
import cv2
import numpy as np
import yaml

from src.detector.YOLO import YOLOv5
from predict_detection import *
```

```python
cfg_path = "./configs/cfg_detection.yml"
with open(cfg_path, 'r') as ymlfile:
    cfg = yaml.safe_load(ymlfile)
```

```python
# Loading parameters model from config
weights = cfg["model"]["weights"]
class_names = cfg["model"]["class_names"]
class_id_names = {class_id: class_name for class_id, class_name in enumerate(class_names)}
input_height, input_width = map(int, cfg["model"]["input_size"].split(','))

# Loading parameters for inference
dir_img = cfg["inference"]["dir_img"]
conf_tresh = cfg["inference"]["confidence_threshold"]
iou_thresh= cfg["inference"]["iou_threshold"]
```

```python
detector = YOLOv5(input_res=(input_height, input_width),
                  weights=weights,
                  batch_size=1)
```

```python
path_image = "./data/test.jpg"
img = cv2.imread(path_image)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
predict_image(img, detector, conf_thresh=conf_tresh, iou_thresh=iou_thresh)
```
